## Training the Convolutional Layer in the ANN

Convolutions
- Use a kernel matrix to scan a given image and apply a filter to obtain a certain effect.
- Preserve the spatial relationship between pixels.

Kernels/Feature Detectors
- An image "Kernel" is a matrix used to apply effects such as blurring and sharpening on the image.
- Are used in ML for Feature Extraction to spelect the most important pixels of an image.

Other Terms to Note
- The Sigmoid Activation Function- is softer on the change from (0,1) since it gives a probability of being either (0,1)

- The Rectified Linear Unit (RELU) Activation Function- is another option and this function returns a 0 for any negative value and a 1 for any positive value

- Pooling- This helps reduce feature map dimensionality while improving computational efficiency without comprimising features. It helps avoid overfitting.

- Flattening- In this case flattening is when the 28x28 image and to flatten it into a single array.

- Dropout Layers- will randomly turn off layer weights to allow the NN to not overfit with the dataset. 

```python
# Prepare the training and testing dataset 
X_train = training[:,1:]/255
y_train = training[:,0]

X_test = testing[:,1:]/255
y_test = testing[:,0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_validate, y_train, y_validate = train_test_split(X_train, y_train, test_size = 0.2, random_state = 12345)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# * unpack the tuple
X_train = X_train.reshape(X_train.shape[0], *(28, 28, 1))
X_test = X_test.reshape(X_test.shape[0], *(28, 28, 1))
X_validate = X_validate.reshape(X_validate.shape[0], *(28, 28, 1))
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_validate.shape
```

```python
# Import train_test_split from scikit library
# Import Keras
import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
cnn_model = Sequential()

# Try 32 fliters first then 64
cnn_model.add(Conv2D(64,3, 3, input_shape = (28,28,1), activation='relu'))
cnn_model.add(MaxPooling2D(pool_size = (2, 2)))

cnn_model.add(Dropout(0.25))

# cnn_model.add(Conv2D(32,3, 3, activation='relu'))
# cnn_model.add(MaxPooling2D(pool_size = (2, 2)))

cnn_model.add(Flatten())
cnn_model.add(Dense(output_dim = 32, activation = 'relu'))
cnn_model.add(Dense(output_dim = 10, activation = 'sigmoid'))
```

```python
cnn_model.compile(loss ='sparse_categorical_crossentropy', optimizer=Adam(lr=0.001),metrics =['accuracy'])
```

```python
epochs = 50 #The number of iterations ex. 50x

history = cnn_model.fit(X_train,
                        y_train,
                        batch_size = 512,
                        nb_epoch = epochs,
                        verbose = 1,
                        validation_data = (X_validate, y_validate))
```

## Evaluating the Model's Performance

```python
evaluation = cnn_model.evaluate(X_test, y_test)
print('Test Accuracy : {:.3f}'.format(evaluation[1]))
```

- Our initial accuracy in kernel 32 was 91.1% which is great since in the test it was 95.1%.
- After switching kernel 32 into kernel 64 and adding dropouts the performance of the model increased to 91.8%.

```python
# Get the predictions for the test data
predicted_classes = cnn_model.predict_classes(X_test)
```
